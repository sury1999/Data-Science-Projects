```
#Importing required libraries
import numpy as np
from skimage import io
from sklearn.cluster import KMeans
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```
#Reading the image data
img  = io.imread('/content/taj_mahal.jpg')
plt.subplots(figsize=(14,6))
plt.imshow(img)
plt.show()
```

```
#Processing image data to work with it. RGB values lie b/w 0 and 255 therefore dividing by 255. 
#Converting into array with 3 columns
img_data = (img/255).reshape(-1, 3)
img_data.shape
```

```
#Using 32 clusters for first image
kmeans_1 = KMeans(n_clusters=32).fit(img_data)
k_colors_1 = kmeans_1.cluster_centers_[kmeans_1.predict(img_data)]
```

```
#Using 16 cluster for second image,
kmeans_2 =  KMeans(n_clusters=16).fit(img_data)
k_colors_2 = kmeans_2.cluster_centers_[kmeans_2.predict(img_data)]
```
